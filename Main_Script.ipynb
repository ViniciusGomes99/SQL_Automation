```python
from mysql import connector
import pandas as pd

# Connect to your MySQL database
db_connection = connector.connect(
    host="localhost",
    user="root",
    password="7040",
    database="acadiatest"
)

cursor = db_connection.cursor()

columns_to_insert = ['DepartmentID', 'Year', 'SegmentID', 'ProfileID', 'Sales', 'Customers']

excel_file = pd.read_excel(r"C:\Users\Vinicius\Desktop\Acadia Test\Acadia - Sr Analyst Case Study.xlsx", sheet_name="S2 - Dept Data")

# Iterate through the rows of the DataFrame and insert data into the MySQL table
for index, row in excel_file[columns_to_insert].iterrows():
    # Convert NaN values to 'NaN' as strings
    row = row.apply(lambda x: str(x) if pd.notna(x) else 'NaN')
    print(row)
    
    data = tuple(row)
    
    insert_query = "INSERT INTO fHistory (DepartmentID, Year, SegmentID, ProfileID, Sales, Customers) VALUES (%s, %s, %s, %s, %s, %s)"
    
    cursor.execute(insert_query, data)

# Commit the changes to the database
db_connection.commit()

# Close the cursor and database connection
cursor.close()
db_connection.close()
```
